```python
import requests
key = 'W1x0upoG8ZmLyh5J5WhbaUhz4mQUC6Nv'
url = 'http://www.mapquestapi.com/geocoding/v1/address?key='
#loc='budigere,bangalore'
def locate(loc):
    main = url+key+'&location='+loc
    r=requests.get(main)
    data = r.json()['results'][0]['locations'][0]['latLng']['lat'],r.json()['results'][0]['locations'][0]['latLng']['lng'],r.json()['results'][0]['locations'][0]['postalCode']
    return data
#print(data)
```

```python
import pandas as pd

df = pd.read_csv('Mumbai.csv')
df.head()
```

```python
dic={}
```

```python
unq = df.Location.unique()
unq[23]
len(unq)
```

```python
for i in unq:
    dic[i] = locate(i)
```

```python
lt=[]
lg=[]
for i in df['Location']:
  lt.append(dic[i][0])
  lg.append(dic[i][1])
```

```python
df['lat'] = lt
df['lon'] = lg
```
